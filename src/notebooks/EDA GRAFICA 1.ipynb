```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white")
import missingno as msno

# dataset original
df =pd.read_csv('../data/ring_unico.csv')
```

```python
# -------------------------------- graficas sin requerimiento de cambios en el dataset original ----------------------------------------------
```

```python
# ganador rojo, azul
fig, ax = plt.subplots(1,1, figsize=(12, 7), dpi=72)
sns.countplot(x="Winner", data=df, ax=ax)
```

```python
# pelea por clases
fig, ax = plt.subplots(1,1, figsize=(12, 7), dpi=72)
ax.set_facecolor('xkcd:black')
fig.patch.set_facecolor('dimgray')
sns.countplot(x="weight_class", data=df, ax=ax, order=df.weight_class.value_counts().iloc[:].index)
plt.xticks(rotation=15)
plt.savefig('clases.png');
```

```python
# peleas por ao
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['Year'] = df['date'].dt.year
fig, ax = plt.subplots(1,1, figsize=(12, 7), dpi=72)
sns.countplot(x="Year", data=df, ax=ax)
plt.xticks(rotation=15)
ax.set_facecolor('xkcd:black')
fig.patch.set_facecolor('orange')
plt.savefig('ao.png');
```
